```python
df = pd.read_excel("FullAvatarDescriptions.xlsx")
df
```

```python
newDf = df[['AvatarId','Description']]
newDf
```

```python
removeIndices = []
for i in df.index:
    avatarId = df.loc[i]['AvatarId']
    #print(f"{type(avatarId)},{avatarId}")
    #print(f"{len(avatarId)},{avatarId[0:7]}\n")
    if type(avatarId) != str or len(avatarId) < 6 or avatarId[0:6] != "avatar":
        removeIndices.append(i)
print(len(removeIndices))
newDf2 = newDf.drop(removeIndices)
newDf2
```

```python
newDf2.to_excel("final_dataset.xlsx")
```
